### K-Nearest Neighbhors


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

#Step #1: Load your data
url = 'https://drive.google.com/file/d/17ftgPDTeaQOcBEI4eDRehk5qSZdFQKAv/view?usp=sharing'
file_id=url.split('/')[-2]
dwn_url='https://drive.google.com/uc?id=' + file_id
df = pd.read_csv(dwn_url)
```

```python
### Converting the data into numerical using one hot encoding
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder(categories=[['sunny', 'overcast', 'rainy']])
encoder2 = OrdinalEncoder(categories=[['hot', 'mild', 'cool']])
encoder3 = OrdinalEncoder(categories=[['high', 'normal']])
encoder4 = OrdinalEncoder(categories=[['True', 'False']])
encoder5 = OrdinalEncoder(categories=[['no', 'yes']])

df['Outlook_encoded'] = encoder.fit_transform(df[['outlook']])
df['temp_encoded'] = encoder2.fit_transform(df[['temp']])
df['humidity_encoded'] = encoder3.fit_transform(df[['humidity']])
df['play_encoded'] = encoder5.fit_transform(df[['play']])
df
```

```python
X = df.drop(['play','outlook','temp','humidity','play_encoded'],axis=1)
y = df['play_encoded']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
